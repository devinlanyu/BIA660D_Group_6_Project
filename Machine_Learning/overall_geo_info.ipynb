```python
import pandas as pd
```

```python
df=pd.read_csv('overall.csv')
df.pop('Unnamed: 0')
df.pop('Unnamed: 8')
geo_df=pd.read_csv('countries.csv')
```

```python
df.head()
```

```python
geo_df.head()
```

```python
lats=[]
lons=[]
for i in range(len(df)):
    temp=geo_df.loc[geo_df['Countries'] ==  df['Nationality'][i]]
    lats.append(float(temp['latitude'].values))
    lons.append(float(temp['longitude'].values))
```

```python
df['latitude']=lats
df['longitude']=lons
df.head()
```

```python
df.to_csv('overall_with_geo.csv')
```

```python
dummy_ranks = pd.get_dummies(df['Nationality'])
```

```python
countries=dummy_ranks.columns.tolist()
count=[]
for i in range(len(dummy_ranks.columns)):
    count.append(sum(dummy_ranks.iloc[:,i]))
```

```python
map_df=pd.DataFrame()
map_df['countries']=countries
map_df['count']=count
map_df.head()
```

```python
lats=[]
lons=[]
for i in range(len(map_df)):
    temp=geo_df.loc[geo_df['Countries'] ==  map_df['countries'][i]]
    lats.append(float(temp['latitude'].values))
    lons.append(float(temp['longitude'].values))
```

```python
map_df['latitude']=lats
map_df['longitude']=lons
```

```python
map_df.to_csv('map_info.csv')
```
