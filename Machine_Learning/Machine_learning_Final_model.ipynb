```python
from sklearn import preprocessing
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split, KFold
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.model_selection import cross_val_score
```

```python
y=df_latest['Score']
DF1=df_latest.drop(['Name','Nationality','Score','friends','solo', 'Date_new'],axis=1)
```

```python
def remove_punctuation(text):
    '''a function for removing punctuation'''
    import string
    # replacing the punctuations with no space, 
    # which in effect deletes the punctuation marks 
    translator = str.maketrans('', '', string.punctuation)
    # return the text stripped of punctuation marks
    return text.translate(translator)
```

```python
sw = stopwords.words('english')
def stopwords(text):
    '''a function for removing the stopword'''
    # removing the stop words and lowercasing the selected words
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    # joining the list of words with space separator
    return " ".join(text)
```

```python
stemmer = SnowballStemmer("english")

def stemming(text):    
    '''a function which stems each word in the given text'''
    text = [stemmer.stem(word) for word in text.split()]
    return " ".join(text)
```

```python
DF1['Topic']=DF1['Topic'].apply(remove_punctuation)
DF1['Topic']=DF1['Topic'].apply(stopwords)
DF1['Topic']=DF1['Topic'].apply(stemming)
```

```python
cv = CountVectorizer()
data = cv.fit_transform(DF1.Topic)
```
