```python
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()

def clean(doc):
    stop_free = ''.join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join([ch for ch in stop_free if ch not in exclude])
    normalized = ''.join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
doc_clean = [clean(doc).split() for doc in list_new]
```

```python
from gensim.models import word2vec
import logging
import pandas as pd
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = word2vec.Word2Vec(doc_clean, min_count=5, size=200, window=5, workers=4 )
```

```python
print("Top 20 words similar to word 'sound'")
print(model.wv.most_similar('room', topn=20))

print("Top 5 words similar to word 'room' but not relevant to 'bathroom'")
print(model.wv.most_similar(positive=['bed'], negative=['value'], topn=10))
```
