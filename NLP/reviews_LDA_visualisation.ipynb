```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
df=pd.read_csv('wynn_pos_reviews.csv')
```

```python
df.columns.values.tolist()
```

```python
df.drop(['Unnamed: 0'], axis=1,inplace=True)
```

```python
corpus=[]
stop_words={'get','though','enjoy','probabl','us','everyth'}
for i in range(570):    
    review=re.sub('[^a-zA-Z]',' ',df['0'][i])
    review=review.lower()
    review=review.split()
    stemmer=PorterStemmer()
    review=[stemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
cv=CountVectorizer(max_features=1000,ngram_range=(2, 2))
x=cv.fit_transform(corpus)
```

```python
lda = LatentDirichletAllocation(n_components=5, max_iter=50,

                                learning_method='online',

                                learning_offset=50.,

                                random_state=0)
```

```python
lda.fit(x)
```

```python
def print_top_words(model, feature_names, n_top_words):

    for topic_idx, topic in enumerate(model.components_):

        print("Topic #%d:" % topic_idx)

        print(" ".join([feature_names[i]

                        for i in topic.argsort()[:-n_top_words - 1:-1]]))

    print()
```

```python
n_top_words = 20
```

```python
cv_feature_names = cv.get_feature_names()

print_top_words(lda, cv_feature_names, n_top_words)
```
